## Задача D
По данным целым неотрицательным n и k вычислите значение числа сочетаний из n элементов по k, то есть $\frac{n!}{k!(n-k)!}$.

```python
n = int(input('n = '))
k = int(input('k = '))
nf = 1
kf = 1
nkf = 1
for i in range(1, n+1):
  nf = nf * i
for i in range(1, k+1):
  kf = kf * i
for i in range(1, n-k+1):
  nkf = nkf * i
c = nf / kf / nkf
print("c = ", c)
```
